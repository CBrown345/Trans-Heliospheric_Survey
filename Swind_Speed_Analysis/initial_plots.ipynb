```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["text.usetex"] = True
```

```python
df_fast = pd.read_hdf(rf"..\Data_Processed\msc\msc_fast_binned.hf")
df_med  = pd.read_hdf(rf"..\Data_Processed\msc\msc_med_binned.hf")
df_slow = pd.read_hdf(rf"..\Data_Processed\msc\msc_slow_binned.hf")
```

```python
d_mercury =  0.3871   # Semi-major axes of the orbits [au]
d_venus   =  0.7233
d_earth   =  1.000
d_mars    =  1.524
d_ceres   =  2.767
d_jupiter =  5.204
d_saturn  =  9.583
d_uranus  = 19.22
d_neptune = 30.11
d_pluto   = 39.48
d_eris    = 67.781

lst_d = [ d_mercury, d_venus, d_earth, d_mars, d_ceres,
            d_jupiter, d_saturn, d_uranus, d_neptune, d_pluto, d_eris ]
```

```python
plt.scatter(df_fast["sc_r_med"], df_fast["vp_m_med"], s = 8)
# plt.scatter(df_med["sc_r_med"], df_med["vp_m_med"], s = 8)
plt.scatter(df_slow["sc_r_med"], df_slow["vp_m_med"], s = 8)
plt.xscale("log")
plt.legend(["Fast Wind","Med Wind", "Slow Wind"])
plt.xlabel(r"$r~\rm [au]$")
plt.ylabel(r"$v_p~\rm [km/s]$")
for p in lst_d:
    plt.axvline(p, c = "black", ls = ":", alpha = .5)
```

```python
plt.scatter(df_fast["sc_r_med"], df_fast["bm_med"], s = 8)
# plt.scatter(df_med["sc_r_med"], df_med["vp_m_med"], s = 8)
plt.scatter(df_slow["sc_r_med"], df_slow["bm_med"], s = 8)
plt.xscale("log")
plt.yscale("log")
plt.legend(["Fast Wind", "Slow Wind"])
#plt.legend(["Fast Wind","Med Wind", "Slow Wind"])
plt.xlabel(r"$r~\rm [au]$")
plt.ylabel(r"$B~\rm [nT]$")
for p in lst_d:
    plt.axvline(p, c = "black", ls = ":", alpha = .5)
```

```python
plt.scatter(df_fast["sc_r_med"], df_fast["Tp_med"], s = 8)
# plt.scatter(df_med["sc_r_med"], df_med["vp_m_med"], s = 8)
plt.scatter(df_slow["sc_r_med"], df_slow["Tp_med"], s = 8)
plt.xscale("log")
plt.yscale("log")
plt.legend(["Fast Wind", "Slow Wind"])
#plt.legend(["Fast Wind","Med Wind", "Slow Wind"])
plt.xlabel(r"$r~\rm [au]$")
plt.ylabel(r"$T_p~\rm [K]$")
for p in lst_d:
    plt.axvline(p, c = "black", ls = ":", alpha = .5)
```

```python
plt.scatter(df_fast["sc_r_med"], df_fast["np_med"], s = 8)
# plt.scatter(df_med["sc_r_med"], df_med["vp_m_med"], s = 8)
plt.scatter(df_slow["sc_r_med"], df_slow["np_med"], s = 8)
plt.xscale("log")
plt.yscale("log")
plt.legend(["Fast Wind", "Slow Wind"])
#plt.legend(["Fast Wind","Med Wind", "Slow Wind"])
plt.xlabel(r"$r~\rm [au]$")
plt.ylabel(r"$n_p~\rm [cm^{-3}]$")
for p in lst_d:
    plt.axvline(p, c = "black", ls = ":", alpha = .5)
```

```python
plt.scatter(df_fast["sc_r_med"], df_fast["vp_m_med"], s = 8)
plt.scatter(df_med["sc_r_med"], df_med["vp_m_med"], s = 8)
plt.scatter(df_slow["sc_r_med"], df_slow["vp_m_med"], s = 8)
plt.vlines(df_fast["sc_r_med"], df_fast["vp_m_low"], df_fast["vp_m_high"], colors = "blue")
plt.vlines(df_med["sc_r_med"], df_med["vp_m_low"], df_med["vp_m_high"], colors = "orange")
plt.vlines(df_slow["sc_r_med"], df_slow["vp_m_low"], df_slow["vp_m_high"], colors = "green")
plt.xscale("log")
plt.legend(["Fast Wind","Med Wind", "Slow Wind"])
plt.xlabel(r"$r~\rm [au]$")
plt.ylabel(r"$v_p~\rm [km/s]$")
for p in lst_d:
    plt.axvline(p, c = "black", ls = ":", alpha = .5)
```
