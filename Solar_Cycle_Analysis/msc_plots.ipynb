```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["text.usetex"] = True
```

```python
scs = [
    "psp",
    "solo",
    "helios1",
    "helios2",
    "mariner2",
    "mariner10",
    "ulysses",
    "cassini",
    "pioneer10",
    "pioneer11",
    "newhorizons",
    "voyager1",
    "voyager2"
]
sc_color = [ 'tab:brown', 'g'       , 'tab:orange', 
           		 'tab:olive', 'w'       , 'k'         ,
                 'tab:gray' ,'tab:green', 'tab:purple', 
                 'tab:pink' ,'tab:red'  , 'tab:blue'  , 
              	 'tab:cyan'    ]

sc_marker = [ 'd', '8', '^', 
              	  'v', 'o', 'P', 
                  'X', 'p', '<',
                  '>', '*', 's', 
                  'D'  ]

sc_markeredgewidth = [ 0, 0  , 0,
                       	   0, 0.5, 0, 
                           0, 0  , 0,
                           0, 0  , 0, 
                           0  ]

sc_markersize = [ 3, 3, 3, 
	                  3, 2, 3,
	                  3, 3, 3,
	                  3, 3, 3,
                      3]

sc_markeredgecolor = [ None, None, None,
	                       None, 'k' , None,
	                       None, None, None,
	                       None, None, None,
                           None]
```

```python
df = pd.read_hdf(f"../Data_Processed/Individual/{scs[0]}/{scs[0]}_binned.hf")
#print(df.keys)
```

```python
for n in range(0, len(scs)):
    #print(scs[n])
    df = pd.read_hdf(f"../Data_Processed/Individual/{scs[n]}/{scs[n]}_binned.hf")
    plt.plot(df["sc_r_iqr_50"], df["bm_iqr_50"], linestyle='', color=sc_color[n],marker=sc_marker[n],
		              markersize=sc_markersize[n],
		              markeredgewidth=sc_markeredgewidth[n],
		              markeredgecolor=sc_markeredgecolor[n] )
plt.yscale("log")
plt.xscale("log")
plt.xlabel(r"$r~\rm [au]$")
plt.ylabel(r"$B~\rm [nT]$")
plt.show()
```
